## Ferramentas

Uma das etapas mais importantes para um bom modelo é o tratamento dos dados. Esse processo visa preparar dados brutos em informações que serão úteis para serem analisadas. Uma das técnicas que utilizamos neste trabalho foi a normalização padrão; essa transformação não altera o quão normalmente distribuídos estão os dados, apenas os translada e “estica/comprime” para que apresentem média 0 e desvio padrão 1. [1] Em nosso trabalho, utilizamos essa técnica em conjunto com o *SelectKBest*, que foca em selecionar as melhores variáveis, ou seja, as mais informativas.[2] Também utilizamos a codificação One-Hot para transformar dados categóricos em dados numéricos, armazenando-os em formato binário.[1]

Para trabalhar com aprendizado de máquina, é necessário entender se o modelo será supervisionado ou não. No nosso caso, o aprendizado foi supervisionado, o que significa que o modelo recebe um ou mais atributos (também conhecidos como features) e retorna a previsão de um ou mais alvos (também conhecidos como targets). Quando o alvo a ser predito é uma variável numérica, dizemos que o problema sendo resolvido é de regressão.[3]

**Algoritmo k-Nearest Neighbors (kNN)**

O k-Nearest Neighbors (kNN) é um algoritmo baseado na hipótese de que dados similares ocupam regiões próximas no espaço de entrada (isto é, no espaço dos atributos).[3] Seu princípio é baseado em uma métrica conhecida, como a distância Euclidiana. O valor predito pelo modelo será a média dos alvos dos  𝑘k vizinhos mais próximos do exemplo de entrada (isto é, os 𝑘k vizinhos com menor distância).[3]

**Algoritmo Árvore de Decisão**

Esse algoritmo utiliza um tipo de grafo direcional acíclico composto de três tipos de vértices ou nós: raiz, decisão e folha. A raiz e os vértices de decisão contêm condicionais, e os vértices folha contêm um valor de resposta. Induzir um modelo de árvore de decisão envolve definir quais serão os condicionais dos vértices raiz e de decisão, e quais são os valores de resposta dos vértices folha.[4] No entanto, apesar de ser um modelo simples, ele pode ter um desempenho inferior ao de outros algoritmos.

**Algoritmo Floresta Aleatória**

Como diria o professor Daniel, “com várias árvores, forma-se uma floresta” – e é por esse caminho que esse algoritmo segue. A floresta aleatória é composta de várias árvores de decisão, sendo "aleatória" porque seu processo de construção envolve amostragem de exemplos e dos atributos.[5]

## Teste Linear Regressor - PCA

Nesta seção, testaremos a performance do modelo após reduzir a dimensionalidade com o uso de PCA (Análise das Componentes Principais ou Principal Component Analysis). De forma simplificada, esta técnica consiste em alterar o sistema de coordenadas dos dados de forma que as novas coordenadas (obtidas após uma transformação linear) estejam ordenadas com relação a variância dos dados [6].

Criamos uma função que retorna a instância de um pipeline que realiza PCA nos dados e treina um modelo de floresta aleatória.

```python
def cria_modelo_linear_regression_pca(num_dimensoes):

    modelo_linear_regression_pca = make_pipeline(
        PCA(num_dimensoes),
        LinearRegression()
    )
    
    return modelo_linear_regression_pca
```

## Teste Linear Regression - Seleção VIF

Realizamos o teste de performance do modelo após realizar seleção de atributos com VIF (fator de inflação de variância ou variance inflation factor). Esse método reduz a multicolinearidade dos dados ao remover colunas que são combinação linear de uma ou mais outras colunas [7].

```python
def selecao_vif(df_atributos, limiar_vif):
    """Realiza a seleção de atributos por VIF.
    Args:
      df_atributos: DataFrame contendo os atributos.
      limiar_vf: valor do limiar do vif. Número positivo. Usualmente é 5 ou 10.
    Returns:
      DataFrame com os atributos selecionados.
      
      Função criada pelo professor Daniel Cassar
    """
    
    df = df_atributos.copy()
    while True:
        VIFs = []
        for col in df.columns:
            X = df.drop(col, axis=1).values
            y = df[col].values
            r_quadrado = LinearRegression().fit(X, y).score(X, y)
            if r_quadrado != 1:
                VIF = 1 / (1 - r_quadrado)
            else:
                VIF = float("inf")
            VIFs.append(VIF)
        VIF_maximo = max(VIFs)
        if VIF_maximo > limiar_vif:
            indice = VIFs.index(VIF_maximo)
            coluna_remocao = df.columns[indice]
            df = df.drop(coluna_remocao, axis=1)
        else:
            break
    return df

X_modificado = selecao_vif(pd.DataFrame(X), limiar_vif=10).values
X_modificado.shape
```

## Teste Linear Regression - Seleção de Atributos e Normalizador Padrão

Nesta sessão usaremos um pipeline que aplica normalização padrão nos dados numéricos e converte os categóricos com One-Hot, além de selecionar os melhores atributos com a função SelectKBest, do scikit-learn.

```python
preprocessor = ColumnTransformer(
    transformers=[('cat', OneHotEncoder(), CATEGORICAL_FEATURES),
                  ('num', StandardScaler(), NUMERIC_FEATURES)]
)

def cria_modelo_floresta_selecao(n_melhores):
    pipeline_selecao_floresta = make_pipeline(
        preprocessor,                             
        SelectKBest(score_func=f_regression, k=n_melhores), 
        LinearRegression()         
    )
    
    return pipeline_selecao_floresta
```

## Otimização de hiperparâmetros df puro

```python
def cria_instancia_KNN(trial):

    parametros = {
        'n_neighbors': trial.suggest_int(
            'n_vizinhos', 3, 15
        ),
        'weights': trial.suggest_categorical(
            'pesos', ['uniform', 'distance']
        ),
        'algorithm': trial.suggest_categorical(
            'algoritimo', ['ball_tree', 'kd_tree', 'brute']
        ),
        'leaf_size': trial.suggest_int(
            'tamanho_folha', 5, 45
        ),
        'p': 1, 
        'n_jobs': -1
    }
    
    model = KNeighborsRegressor(**parametros)
    
    return model
```

```python
def funcao_objetivo_KNN(trial, X, y, num_folds):

    modelo = cria_instancia_KNN(trial)

    metricas = cross_val_score(
        modelo,
        X,
        y,
        scoring="neg_root_mean_squared_error",
        cv=num_folds,
    )

    return -metricas.mean()
```

## Teste K-NN - SELEÇÃO VIF:

```python
def selecao_vif(df_atributos, limiar_vif):
    
    df = df_atributos.copy()
    while True:
        VIFs = []
        for col in df.columns:
            X = df.drop(col, axis=1).values
            y = df[col].values
            r_quadrado = LinearRegression().fit(X, y).score(X, y)
            if r_quadrado != 1:
                VIF = 1 / (1 - r_quadrado)
            else:
                VIF = float("inf")
            VIFs.append(VIF)
        VIF_maximo = max(VIFs)
        if VIF_maximo > limiar_vif:
            indice = VIFs.index(VIF_maximo)
            coluna_remocao = df.columns[indice]
            df = df.drop(coluna_remocao, axis=1)
        else:
            break
    return df

X_modificado = selecao_vif(pd.DataFrame(X), limiar_vif=10).values
X_modificado.shape
```

### Otimização de hiperparâmetros - df puro

Nesta seção, realizamos a otimização dos hiperparâmetros do modelo com a biblioteca Optuna.

Criamos uma função em que definimos os hiperparâmetros a serem testados.

```python
def cria_instancia_floresta(trial):

    parametros = {
        'n_estimators': trial.suggest_int(
            'n_arvores', 5, 400
        ),
        'min_samples_split': trial.suggest_float(
            'min_exemplos_split', 0, 1
        ),
        'min_samples_leaf': trial.suggest_float(
            'min_exemplos_folha', 0, 1
        ),
        'max_features': trial.suggest_float(
            'max_atributos', 0, 1
        ),
        'max_samples': trial.suggest_float(
            'max_exemplos', 0, 1
        ),
        'random_state': SEMENTE_ALEATORIA,
        'bootstrap': True,
        'n_jobs':-1
    }

    modelo = RandomForestRegressor(**parametros)

    return modelo
```

Criamos a função objetivo do Optuna e definimos como testar o modelo (validação cruzada) e a métrica (RMSE).

```python
def funcao_objetivo_floresta(trial, X, y, num_folds):

    modelo = cria_instancia_floresta(trial)

    metricas = cross_val_score(
        modelo,
        X,
        y,
        scoring="neg_root_mean_squared_error",
        cv=num_folds,
    )

    return -metricas.mean()
```

Criamos o objeto de estudo para realizar a otimização.

```python
NOME_DO_ESTUDO = "teste_floresta_df_puro"

objeto_de_estudo = create_study(
    direction="minimize",
    study_name=NOME_DO_ESTUDO,
    storage=f"sqlite:///{NOME_DO_ESTUDO}.db",
    load_if_exists=True,
)
```

```python
def funcao_objetivo_parcial_floresta(trial):
    return funcao_objetivo_floresta(trial, X_treino, y_treino, NUM_FOLDS)
```

Realizamos a otimização com o optuna, testando 300 conjuntos de hiperparâmetros.

```python
NUM_TENTATIVAS = 300

objeto_de_estudo.optimize(funcao_objetivo_parcial_floresta, n_trials=NUM_TENTATIVAS)
```

Carregamos o objeto de estudo e guardamos a melhor trial.
