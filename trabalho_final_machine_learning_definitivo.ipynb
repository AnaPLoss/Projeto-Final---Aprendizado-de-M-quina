## Ferramentas

Uma das etapas mais importantes para um bom modelo √© o tratamento dos dados. Esse processo visa preparar dados brutos em informa√ß√µes que ser√£o √∫teis para serem analisadas. Uma das t√©cnicas que utilizamos neste trabalho foi a normaliza√ß√£o padr√£o; essa transforma√ß√£o n√£o altera o qu√£o normalmente distribu√≠dos est√£o os dados, apenas os translada e ‚Äúestica/comprime‚Äù para que apresentem m√©dia 0 e desvio padr√£o 1. [1] Em nosso trabalho, utilizamos essa t√©cnica em conjunto com o *SelectKBest*, que foca em selecionar as melhores vari√°veis, ou seja, as mais informativas.[2] Tamb√©m utilizamos a codifica√ß√£o One-Hot para transformar dados categ√≥ricos em dados num√©ricos, armazenando-os em formato bin√°rio.[1]

Para trabalhar com aprendizado de m√°quina, √© necess√°rio entender se o modelo ser√° supervisionado ou n√£o. No nosso caso, o aprendizado foi supervisionado, o que significa que o modelo recebe um ou mais atributos (tamb√©m conhecidos como features) e retorna a previs√£o de um ou mais alvos (tamb√©m conhecidos como targets). Quando o alvo a ser predito √© uma vari√°vel num√©rica, dizemos que o problema sendo resolvido √© de regress√£o.[3]

**Algoritmo k-Nearest Neighbors (kNN)**

O k-Nearest Neighbors (kNN) √© um algoritmo baseado na hip√≥tese de que dados similares ocupam regi√µes pr√≥ximas no espa√ßo de entrada (isto √©, no espa√ßo dos atributos).[3] Seu princ√≠pio √© baseado em uma m√©trica conhecida, como a dist√¢ncia Euclidiana. O valor predito pelo modelo ser√° a m√©dia dos alvos dos  ùëòk vizinhos mais pr√≥ximos do exemplo de entrada (isto √©, os ùëòk vizinhos com menor dist√¢ncia).[3]

**Algoritmo √Årvore de Decis√£o**

Esse algoritmo utiliza um tipo de grafo direcional ac√≠clico composto de tr√™s tipos de v√©rtices ou n√≥s: raiz, decis√£o e folha. A raiz e os v√©rtices de decis√£o cont√™m condicionais, e os v√©rtices folha cont√™m um valor de resposta. Induzir um modelo de √°rvore de decis√£o envolve definir quais ser√£o os condicionais dos v√©rtices raiz e de decis√£o, e quais s√£o os valores de resposta dos v√©rtices folha.[4] No entanto, apesar de ser um modelo simples, ele pode ter um desempenho inferior ao de outros algoritmos.

**Algoritmo Floresta Aleat√≥ria**

Como diria o professor Daniel, ‚Äúcom v√°rias √°rvores, forma-se uma floresta‚Äù ‚Äì e √© por esse caminho que esse algoritmo segue. A floresta aleat√≥ria √© composta de v√°rias √°rvores de decis√£o, sendo "aleat√≥ria" porque seu processo de constru√ß√£o envolve amostragem de exemplos e dos atributos.[5]

## Teste Linear Regressor - PCA

Nesta se√ß√£o, testaremos a performance do modelo ap√≥s reduzir a dimensionalidade com o uso de PCA (An√°lise das Componentes Principais ou Principal Component Analysis). De forma simplificada, esta t√©cnica consiste em alterar o sistema de coordenadas dos dados de forma que as novas coordenadas (obtidas ap√≥s uma transforma√ß√£o linear) estejam ordenadas com rela√ß√£o a vari√¢ncia dos dados [6].

Criamos uma fun√ß√£o que retorna a inst√¢ncia de um pipeline que realiza PCA nos dados e treina um modelo de floresta aleat√≥ria.

```python
def cria_modelo_linear_regression_pca(num_dimensoes):

    modelo_linear_regression_pca = make_pipeline(
        PCA(num_dimensoes),
        LinearRegression()
    )
    
    return modelo_linear_regression_pca
```

## Teste Linear Regression - Sele√ß√£o VIF

Realizamos o teste de performance do modelo ap√≥s realizar sele√ß√£o de atributos com VIF (fator de infla√ß√£o de vari√¢ncia ou variance inflation factor). Esse m√©todo reduz a multicolinearidade dos dados ao remover colunas que s√£o combina√ß√£o linear de uma ou mais outras colunas [7].

```python
def selecao_vif(df_atributos, limiar_vif):
    """Realiza a sele√ß√£o de atributos por VIF.
    Args:
      df_atributos: DataFrame contendo os atributos.
      limiar_vf: valor do limiar do vif. N√∫mero positivo. Usualmente √© 5 ou 10.
    Returns:
      DataFrame com os atributos selecionados.
      
      Fun√ß√£o criada pelo professor Daniel Cassar
    """
    
    df = df_atributos.copy()
    while True:
        VIFs = []
        for col in df.columns:
            X = df.drop(col, axis=1).values
            y = df[col].values
            r_quadrado = LinearRegression().fit(X, y).score(X, y)
            if r_quadrado != 1:
                VIF = 1 / (1 - r_quadrado)
            else:
                VIF = float("inf")
            VIFs.append(VIF)
        VIF_maximo = max(VIFs)
        if VIF_maximo > limiar_vif:
            indice = VIFs.index(VIF_maximo)
            coluna_remocao = df.columns[indice]
            df = df.drop(coluna_remocao, axis=1)
        else:
            break
    return df

X_modificado = selecao_vif(pd.DataFrame(X), limiar_vif=10).values
X_modificado.shape
```

Ap√≥s reduzir o n√∫mero de atributos para 57, treinamos um modelo e o avaliamos com valida√ß√£o cruzada.

```python
metricas = -cross_val_score(
    LinearRegression(),
    X_modificado,
    y,
    scoring="neg_root_mean_squared_error",
    cv=NUM_FOLDS,
    n_jobs=-1
)

print(f"M√©tricas: {metricas}\n")
print(f"M√©dia das m√©tricas: {metricas.mean()}\n")
```

N√£o se observa mudan√ßa na performance do modelo ao aplicar VIF.

## Teste Linear Regression - Sele√ß√£o de Atributos e Normalizador Padr√£o

Nesta sess√£o usaremos um pipeline que aplica normaliza√ß√£o padr√£o nos dados num√©ricos e converte os categ√≥ricos com One-Hot, al√©m de selecionar os melhores atributos com a fun√ß√£o SelectKBest, do scikit-learn.

```python
preprocessor = ColumnTransformer(
    transformers=[('cat', OneHotEncoder(), CATEGORICAL_FEATURES),
                  ('num', StandardScaler(), NUMERIC_FEATURES)]
)

def cria_modelo_floresta_selecao(n_melhores):
    pipeline_selecao_floresta = make_pipeline(
        preprocessor,                             
        SelectKBest(score_func=f_regression, k=n_melhores), 
        LinearRegression()         
    )
    
    return pipeline_selecao_floresta
```

```python
def cria_instancia_KNN(trial):

    parametros = {
        'n_neighbors': trial.suggest_int(
            'n_vizinhos', 3, 15
        ),
        'weights': trial.suggest_categorical(
            'pesos', ['uniform', 'distance']
        ),
        'algorithm': trial.suggest_categorical(
            'algoritimo', ['ball_tree', 'kd_tree', 'brute']
        ),
        'leaf_size': trial.suggest_int(
            'tamanho_folha', 5, 45
        ),
        'p': 1, 
        'n_jobs': -1
    }
    
    model = KNeighborsRegressor(**parametros)
    
    return model
```

## Teste K-NN - SELE√á√ÉO VIF:

```python
def selecao_vif(df_atributos, limiar_vif):
    
    df = df_atributos.copy()
    while True:
        VIFs = []
        for col in df.columns:
            X = df.drop(col, axis=1).values
            y = df[col].values
            r_quadrado = LinearRegression().fit(X, y).score(X, y)
            if r_quadrado != 1:
                VIF = 1 / (1 - r_quadrado)
            else:
                VIF = float("inf")
            VIFs.append(VIF)
        VIF_maximo = max(VIFs)
        if VIF_maximo > limiar_vif:
            indice = VIFs.index(VIF_maximo)
            coluna_remocao = df.columns[indice]
            df = df.drop(coluna_remocao, axis=1)
        else:
            break
    return df

X_modificado = selecao_vif(pd.DataFrame(X), limiar_vif=10).values
X_modificado.shape
```

```python
def cria_instancia_arvore_decisao(trial):

    parametros = {
        "criterion": trial.suggest_categorical(
            "criterio", ["squared_error", "friedman_mse", "absolute_error", "poisson"]
        ),
        "splitter": trial.suggest_categorical(
            "divisor", ["best", "random"]
        ),
        "max_depth": trial.suggest_int("profundidade_max", 10, 1000),
        "min_samples_split": trial.suggest_float(
            "min_exemplos_split", 0, 1
        ),
        "min_samples_leaf": trial.suggest_float(
            "min_exemplos_folha", 0, 1
        ),
        "max_features": trial.suggest_float("num_max_atributos", 0, 1),
        "random_state": SEMENTE_ALEATORIA,
    }

    model = DecisionTreeRegressor(**parametros)

    return model
```

### Otimiza√ß√£o de hiperpar√¢metros - df puro

Nesta se√ß√£o, realizamos a otimiza√ß√£o dos hiperpar√¢metros do modelo com a biblioteca Optuna.

Criamos uma fun√ß√£o em que definimos os hiperpar√¢metros a serem testados.

```python
def cria_instancia_floresta(trial):

    parametros = {
        'n_estimators': trial.suggest_int(
            'n_arvores', 5, 400
        ),
        'min_samples_split': trial.suggest_float(
            'min_exemplos_split', 0, 1
        ),
        'min_samples_leaf': trial.suggest_float(
            'min_exemplos_folha', 0, 1
        ),
        'max_features': trial.suggest_float(
            'max_atributos', 0, 1
        ),
        'max_samples': trial.suggest_float(
            'max_exemplos', 0, 1
        ),
        'random_state': SEMENTE_ALEATORIA,
        'bootstrap': True,
        'n_jobs':-1
    }

    modelo = RandomForestRegressor(**parametros)

    return modelo
```

Criamos a fun√ß√£o objetivo do Optuna e definimos como testar o modelo (valida√ß√£o cruzada) e a m√©trica (RMSE).

```python
def funcao_objetivo_floresta(trial, X, y, num_folds):

    modelo = cria_instancia_floresta(trial)

    metricas = cross_val_score(
        modelo,
        X,
        y,
        scoring="neg_root_mean_squared_error",
        cv=num_folds,
    )

    return -metricas.mean()
```

Criamos o objeto de estudo para realizar a otimiza√ß√£o.

```python
NOME_DO_ESTUDO = "teste_floresta_df_puro"

objeto_de_estudo = create_study(
    direction="minimize",
    study_name=NOME_DO_ESTUDO,
    storage=f"sqlite:///{NOME_DO_ESTUDO}.db",
    load_if_exists=True,
)
```

```python
def funcao_objetivo_parcial_floresta(trial):
    return funcao_objetivo_floresta(trial, X_treino, y_treino, NUM_FOLDS)
```

Realizamos a otimiza√ß√£o com o optuna, testando 300 conjuntos de hiperpar√¢metros.

```python
NUM_TENTATIVAS = 300

objeto_de_estudo.optimize(funcao_objetivo_parcial_floresta, n_trials=NUM_TENTATIVAS)
```

Carregamos o objeto de estudo e guardamos a melhor trial.

De forma simplificada, esta t√©cnica consiste em alterar o sistema de coordenadas dos dados de forma que as novas coordenadas (obtidas ap√≥s uma transforma√ß√£o linear) estejam ordenadas com rela√ß√£o a vari√¢ncia dos dados [6].

Criamos uma fun√ß√£o que retorna a inst√¢ncia de um pipeline que realiza PCA nos dados e treina um modelo de floresta aleat√≥ria.

```python
def cria_modelo_floresta_pca(num_dimensoes):

    modelo_floresta_pca = make_pipeline(
        PCA(num_dimensoes),
        RandomForestRegressor(SEMENTE_ALEATORIA)
    )
    
    return modelo_floresta_pca
```

Testamos reduzir a dimensionalidade de 1 at√© o n√∫mero de atributos menos 1.

```python
for i in range(1, X.shape[1]):

    metricas = -cross_val_score(
        cria_modelo_floresta_pca(i),
        X,
        y,
        scoring="neg_root_mean_squared_error",
        cv=NUM_FOLDS,
        n_jobs=-1
    )

    print(f"M√©tricas com {i} dimensoes: {metricas}\n")
    print(f"M√©dia das m√©tricas com {i} dimensoes: {metricas.mean()}\n")
    print()
```

Nota-se uma perda de performance com at√© 6 dimens√µes. Acima disso, n√£o existe diferen√ßa significativa entre realizar ou n√£o o PCA.

## Teste Floresta Aleat√≥ria - Sele√ß√£o de Atributos e Normalizador Padr√£o

Nesta sess√£o usaremos um pipeline que aplica normaliza√ß√£o padr√£o nos dados num√©ricos e converte os categ√≥ricos com One-Hot, al√©m de selecionar os melhores atributos com a fun√ß√£o SelectKBest, do scikit-learn.

```python
preprocessor = ColumnTransformer(
    transformers=[
        ('cat', OneHotEncoder(), CATEGORICAL_FEATURES),
        ('num', StandardScaler(), NUMERIC_FEATURES)
    ]
)

def cria_modelo_floresta_selecao(n_melhores):
    pipeline_selecao_floresta = make_pipeline(
        preprocessor,                             
        SelectKBest(score_func=f_regression, k=n_melhores), 
        RandomForestRegressor(random_state=SEMENTE_ALEATORIA)         
    )
    
    return pipeline_selecao_floresta
```
